```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

bulan = ['Januari', 'Februaru', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']

#Pengeluaran antara 500.000 - 2.000.000

minggu1 = np.random.randint(500000,2000001,12)
minggu2 = np.random.randint(500000,2000001,12)
minggu3 = np.random.randint(500000,2000001,12)
minggu4 = np.random.randint(500000,2000001,12)

data = {
    'Bulan' : bulan,
    'minggu1' : minggu1,
    'minggu2' : minggu2,
    'minggu3' : minggu3,
    'minggu4' : minggu4
}

df = pd.DataFrame(data)

df['Total Pengeluaran'] = df[['minggu1', 'minggu2', 'minggu3', 'minggu4']].sum(axis=1)
df
```

```python
#Total pengelaran perbulan

plt.figure(figsize=(12,6)) #Canvas dengan ukuran Width, height (inch)
plt.title('Total pengeluaran per bulan')
plt.bar(df['Bulan'], df['Total Pengeluaran'], color='pink')
plt.xticks(rotation=45)
plt.ylabel("Rupiah")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#Scatter plot untuk analisa data pengeluaran per minggu dan per bulan
plt.figure(figsize=(12,6))
for minggu in ['minggu1','minggu2','minggu3','minggu4']:
    plt.plot(df['Bulan'], df[minggu], marker='o', label=minggu)

plt.title("Pengeluaran Per Minggu")
plt.xlabel('Bulan')
plt.ylabel('Pengeluaran (Rupiah)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9,9))
plt.pie(df['Total Pengeluaran'], labels=df['Bulan'], startangle=90, autopct='%1.1f%%')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

df['Bulan_Angka'] = np.arange(1,13)

x = df[['Bulan_Angka']]
y = df['Total Pengeluaran']

model = LinearRegression()

model.fit(x,y)
prediksi = model.predict([[13]])[0]
print(f"Hasil Prediksi adalah: {prediksi}")
```
