```python
import numpy as np
import pandas as pd

data = {
    'Umur': np.random.normal(loc=35, scale=10, size=1000).astype(int).clip(18, 60),
    'Gaji': np.random.lognormal(mean=12, sigma=0.4, size=1000).astype(int).clip(50000, 10000000),
    'Pengalaman': np.random.binomial(n=15, p=0.3, size=1000) + 1,
    'Tingkat Pendidikan': np.random.choice(
        ['SD', 'SMP', 'SMA', 'Sarjana', 'Magister', 'Doktor'],
        size=1000,
        p=[0.05, 0.1, 0.3, 0.4, 0.1, 0.05]
    ),
    'Gender': np.random.choice(
        ['Laki-laki', 'Perempuan'],
        size=1000,
        p=[0.55, 0.45]
    ),
    'Status': np.random.choice(
        ['Belum Menikah', 'Menikah', 'Cerai'],
        size=1000,
        p=[0.4, 0.5, 0.1]
    ),
    'Jumlah Pembelian': np.random.poisson(lam=50, size=1000).clip(1, 1000),
    'Kategori Produk': np.random.choice(
        ['Elektronik', 'Pakaian', 'Furniture', 'Makanan', 'Kesehatan'],
        size=1000,
        p=[0.3, 0.25, 0.2, 0.15, 0.1]
    ),
    'Kepuasan Pelanggan': np.random.choice(
        [1, 2, 3, 4, 5],
        size=1000,
        p=[0.05, 0.1, 0.2, 0.35, 0.3]
    ),
    'Wilayah': np.random.choice(
        ['Jakarta', 'Bandung', 'Bogor', 'Sukabumi', 'Bekasi', 'Depok', 'Tangerang'],
        size=1000,
        p=[0.4, 0.2, 0.1, 0.05, 0.15, 0.05, 0.05]
    )
}

df = pd.DataFrame(data)

print("Contoh Data:")
print(df.head())

print("\nDeskriptif Statistik Numerik:")
print(df.describe())

print("\nDistribusi Kategorikal:")
print("Tingkat Pendidikan:")
print(df['Tingkat Pendidikan'].value_counts())

print("\nGender:")
print(df['Gender'].value_counts())

print("\nStatus:")
print(df['Status'].value_counts())

print("\nKategori Produk:")
print(df['Kategori Produk'].value_counts())

print("\nKepuasan Pelanggan:")
print(df['Kepuasan Pelanggan'].value_counts())

print("\nWilayah:")
print(df['Wilayah'].value_counts())
```
